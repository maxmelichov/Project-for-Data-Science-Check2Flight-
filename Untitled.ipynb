```python
R=pd.DataFrame()
R['R']=df['R']
```

```python
R.drop_duplicates(inplace=True)
R.reset_index(inplace=True)
R.drop('level_0',inplace =True,axis=1)
R.drop('index',inplace =True,axis=1)
print(R)
```

```python
R['count']=0
```

```python
for i in range(len(df['R'])):
    for j in range(len(R['R'])):
        if df['R'][i]==R['R'][j]:
            R['count'][j]+=1
```

```python
R.value_counts(sort=True)
```

```python
R=R.sort_values('R')
```

```python
R
```

```python
R['change_count']=0
```

```python
for i in range(len(R['R'])):
    if i !=22:
        R['change_count'][i]=i*2+abs(500/(22-i))
    else:
         R['change_count'][i]=i*2+abs(500/(1))
```

```python
R
```

```python
R['change_count'].sum()
```

```python
for i in range(len(df['R'])):
    for j in range(len(R['R'])):
        if df['R'][i]==R['R'][j] and R['change_count'][j] > 0:
            df['status'][i]='cancelled'
            R['change_count'][j]-=1
```
